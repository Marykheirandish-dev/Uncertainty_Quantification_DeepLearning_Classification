```python
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
X, y = make_classification(n_samples=7000, n_features=4, n_classes=2, n_clusters_per_class=1,
                           n_informative=4, n_redundant=0,flip_y=0.01 , class_sep=2.5,random_state=42)

# Plot the generated dataset
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', marker='o', edgecolors='k')
plt.title('Random Classification Dataset')
plt.xlabel('X1')
plt.ylabel('X2')
plt.colorbar(label='Class')
plt.grid(True)
plt.show()
```

```python
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
X, y = make_classification(n_samples=7000, n_features=4, n_classes=2, n_clusters_per_class=1,
                           n_informative=4, n_redundant=0,flip_y=0.01 , class_sep=2.5,weights=[0.9, 0.1],random_state=42)

# Plot the generated dataset
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', marker='o', edgecolors='k')
plt.title('Imbalanced Random Classification Dataset')
plt.xlabel('X1')
plt.ylabel('X2')
plt.colorbar(label='Class')
plt.grid(True)
plt.show()
```

```python
# prompt: save X and y as a dataset in a csv file

import csv

with open('dataset1.csv', 'w', newline='') as file:
  writer = csv.writer(file)
  for i in range(len(X)):
    writer.writerow(X[i].tolist() + [y[i]])
```

```python
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt

# Generate an imbalanced dataset
X, y = make_classification(
    n_samples=7000, n_features=5, n_classes=2, n_clusters_per_class=1,
    n_informative=4, n_redundant=0, flip_y=0.01, class_sep=5,
    weights=[0.9, 0.1],  # 90% of class 0, 10% of class 1
    random_state=42
)

# Plot the generated dataset
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', marker='o', edgecolors='k')
plt.title('Imbalanced Classification Dataset')
plt.xlabel('X1')
plt.ylabel('X2')
plt.colorbar(label='Class')
plt.grid(True)
plt.show()
```

```python

import csv

with open('dataset8.csv', 'w', newline='') as file:
  writer = csv.writer(file)
  for i in range(len(X)):
    writer.writerow(X[i].tolist() + [y[i]])
```
